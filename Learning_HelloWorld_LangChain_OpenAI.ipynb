**Create a Simple LangChain "Hello World" Program**

LangChain helps to chain prompts and other language model interactions together. For this simple example, we will create a basic prompt that returns "Hello World!" using OpenAI's language model.

```python
# Create a prompt template
template = "Refine the below draft. Draft: {draft}"

prompt = PromptTemplate(
    input_variables=["draft"],
    template=template,
)

# Initialize the ChatOpenAI language model
llm = ChatOpenAI(model="gpt-4o-mini")  # You can also use "gpt-4" -> change to gpt-4o-mini

# Chat models require a list of messages, not a simple string
messages = [{"role": "system", "content": "You are an product manager. You will send an email to your peer with a good structure language. The email should be in English and make it more easier to understand."},
            {"role": "user", "content": prompt.format(draft="Halo halo, Satu hal lagi, jika kita ingin menambahkan load testing untuk proyek ini, kira-kira berapa effort mandays dan timelinenya. Thank you.")}]

# Invoke the model and get the response
response = llm.invoke(messages)

print(response)
```
